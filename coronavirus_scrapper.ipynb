```python
import requests
page=requests.get("https://www.worldometers.info/coronavirus/#countries")
page
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
list(soup.children)
```

```python
corvid= soup.find(id='main_table_countries')
print(corvid.prettify())
```

```python
actuals=corvid.find('tbody')
print(actuals)
```

```python
A=[]
B=[]
D=[]
E=[]
F=[]
G=[]
H=[]
I=[]

for row in actuals.find_all("tr"):
    cells=row.find_all('td')
    if len(cells)==8:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))
        H.append(cells[6].find(text=True)) 
        I.append(cells[7].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A, columns=['Country, Other'])
#df['Countries,Other']=A
df['Total Cases']=B
df['New Cases']=D
df['Total Deaths']=E
df['New Deaths']=F
df['Active Cases']=G
df['Total Recovered']=H
df['Serious/Critical']=I
df
```

```python
df.to_excel('fafafa.xlsx')
```
